```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def read_point(file_name) :
    li = []
    with open (f'{file_name}.txt', 'r') as f :
        for line in f :
            xyPoint = line.split(' ')
            li.append([])
            x = xyPoint[0]
            y = xyPoint[1][:-1]
            li[-1].append(float(x))
            li[-1].append(float(y))
    return li
```

```python
def read_answer(file_name) :
    li = []
    with open (f'{file_name}.txt', 'r') as f :
        for line in f :
            xyPoint = line.split(' ')
            li.append([])
            x = xyPoint[0]
            y = xyPoint[1][:-1]
            li[-1].append(float(x))
            li[-1].append(float(y))
    return li
```

```python
def draw_coordinate(_list, _c = 'black', _s = 5) :
    xli = []
    yli = []
    for i in range(len(_list)) :
        xli.append(_list[i][0])
        yli.append(_list[i][1])
    plt.scatter(xli, yli, c= _c, s= _s)
```

```python
def dist(p1, p2) :
    
    a1 = p1[0] - p2[0]
    b1 = p1[1] - p2[1]
    
    return math.sqrt(pow(a1,2) + (pow(b1,2)))
```

```python
def draw_2d_line(_list, _color = 'black' ,_line_style = 'solid', _line_width = 0.5) :
    """ draw line.
    Args:
        _list ([[x,y]]): [[x,y]] list. x is start point, y is end point.
        _color (str, optional): line color. Defaults to 'black'.
        _line_style (str, optional): line style. Defaults to 'solid'.
        _line_width (float, optional): line width. Defaults to 0.5.
    """
    for i in range(len(_list)-1) :
        start_x = _list[i][0]
        start_y = _list[i][1]
        end_x = _list[i+1][0]
        end_y = _list[i+1][1]
        plt.plot([start_x, end_x], [start_y, end_y], color = _color, lw = _line_width)
        
    start_x = _list[-1][0]
    start_y = _list[-1][1]
    end_x = _list[0][0]
    end_y = _list[0][1]
    plt.plot([start_x, end_x], [start_y, end_y], color = _color, lw = _line_width)
```
